def func(a, b=6, c=8):
print(a, b, c)
func(1, 2)

3. How about this code: what is its result, and why?
def func(a, *pargs):
    print(a, pargs)
func(1, 2, 3)

output: 
1,(2,3)
# This code defines a function func that takes at least one argument a, followed by any number of additional positional arguments. When the function is called with arguments 1, 2, and 3, 1 is assigned to a, and (2, 3) is packed into a tuple and assigned to pargs. The function then prints out 1 and the tuple (2, 3).
