```python
# 1.	Write a Python program to find words which are greater than given length k?
string = input("Enter a string: ")
k = int(input("Enter a length: "))
result = []

for word in string.split():
    if len(word) > k:
        result.append(word)

print("Words greater than length", k, ":", result)

# 2.	Write a Python program for removing i-th character from a string?
def remove_char(string, i):
    return string[:i] + string[i+1:]

input_string = input("Enter a string: ")
i = int(input("Enter the i-th position of the character to be removed from the string "))
output_string = remove_char(input_string, i)
print(f"Original string: {input_string}")
print(f"Output string: {output_string}")

# 3.	Write a Python program to split and join a string?
strig  ="example thes"
test = strig.split()
" ".join(test)

# 4.	Write a Python to check if a given string is binary string or not?
def check_binary(binary_string):
    for char in binary_string:
        if char not in ['0','1']:
            return False
        return True
binary_string = "10111010110"

if check_binary(binary_string):
    print("Binary string")
else:
    print("not a binary string")
    

# 5.	Write a Python program to find uncommon words from two Strings?

def uncommon_words(str1, str2):
    words1 = str1.split()
    words2 = str2.split()

    set1 = set(words1)
    set2 = set(words2)

    uncommon = set1.symmetric_difference(set2)

    return list(uncommon)


str1 = "this is test string"
str2 = "we are test this line"
print(uncommon_words(str1, str2)) 


# 6.	Write a Python to find all duplicate characters in string?

string = input("Enter a string: ")

test = set()
for char in string:
    if string.count(char) >1:
        test.add(char)
print("Duplicate characters in the string:", test)


# 7.	Write a Python Program to check if a string contains any special character?

import re

def has_special_char(string):
    regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]')
    return regex.search(string) is not None


string = "Hello@World!"

print(has_special_char(string)) 

```
